**n=2  درایه های قطری مثبت و**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[2, 0], [0, 3]])

# Generate points for the ellipse
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

# Scale points to fit the ellipse equation
x_ellipse = x / np.sqrt(A[0, 0])
y_ellipse = y / np.sqrt(A[1, 1])

# Plot the ellipse
plt.figure()
plt.plot(x_ellipse, y_ellipse)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Ellipse: $2x_1^2 + 3x_2^2 = 1$')
plt.axis('equal')
plt.grid(True)
plt.show()
```

**n=3  درایه های قطری مثبت و**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the matrix A
A = np.array([[2, 0, 0], [0, 3, 0], [0, 0, 4]])

# Generate points for the ellipsoid
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# Scale points to fit the ellipsoid equation
x_ellipsoid = x / np.sqrt(A[0, 0])
y_ellipsoid = y / np.sqrt(A[1, 1])
z_ellipsoid = z / np.sqrt(A[2, 2])

# Plot the ellipsoid
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_ellipsoid, y_ellipsoid, z_ellipsoid, color='b', alpha=0.6)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')
ax.set_title('Ellipsoid: $2x_1^2 + 3x_2^2 + 4x_3^2 = 1$')
plt.show()
```

**n=2  درایه های قطری منفی و**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[2, 0], [0, -3]])

# Generate points for the hyperbola
theta = np.linspace(0, 2 * np.pi, 100)
x = np.cosh(theta)
y = np.sinh(theta)

# Scale points to fit the hyperbola equation
x_hyperbola = x / np.sqrt(A[0, 0])
y_hyperbola = y / np.sqrt(-A[1, 1])

# Plot the hyperbola
plt.figure()
plt.plot(x_hyperbola, y_hyperbola, label='Branch 1')
plt.plot(-x_hyperbola, -y_hyperbola, label='Branch 2')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Hyperbola: $2x_1^2 - 3x_2^2 = 1$')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

**n=3  درایه های قطری منفی و**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the matrix A
A = np.array([[2, 0, 0], [0, -3, 0], [0, 0, -4]])

# Generate points for the hyperboloid
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-np.pi / 2, np.pi / 2, 100)
x = np.outer(np.cosh(v), np.cos(u))
y = np.outer(np.sinh(v), np.sin(u))
z = np.outer(np.sinh(v), np.ones_like(u))

# Scale points to fit the hyperboloid equation
x_hyperboloid = x / np.sqrt(A[0, 0])
y_hyperboloid = y / np.sqrt(-A[1, 1])
z_hyperboloid = z / np.sqrt(-A[2, 2])

# Plot the hyperboloid
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_hyperboloid, y_hyperboloid, z_hyperboloid, color='r', alpha=0.6)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$x_3$')
ax.set_title('Hyperboloid: $2x_1^2 - 3x_2^2 - 4x_3^2 = 1$')
plt.show()
```
