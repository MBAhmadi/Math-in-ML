```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# تعریف بردار x و ماتریس A
x_values = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_values, x_values)

# ماتریس‌های مختلف A
A_positive = np.array([[1, 0], [0, 1]])  # ماتریس مثبت معین
A_negative = np.array([[-1, 0], [0, -1]])  # ماتریس منفی معین
A_semi_definite = np.array([[1, 0], [0, 0]])  # نیمه‌معین

# تعریف تابع f(x) = x^T A x
def quadratic_form(A, X, Y):
    return A[0, 0] * X**2 + A[1, 1] * Y**2 + 2 * A[0, 1] * X * Y

# محاسبه خروجی تابع به ازای ماتریس‌های مختلف
Z_positive = quadratic_form(A_positive, X, Y)
Z_negative = quadratic_form(A_negative, X, Y)
Z_semi_definite = quadratic_form(A_semi_definite, X, Y)

# رسم نمودارها
fig = plt.figure(figsize=(18, 6))

# نمودار برای ماتریس مثبت معین
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z_positive, cmap='viridis')
ax1.set_title('Positive Definite A')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x)')

# نمودار برای ماتریس منفی معین
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, Z_negative, cmap='plasma')
ax2.set_title('Negative Definite A')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('f(x)')

# نمودار برای ماتریس نیمه معین
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X, Y, Z_semi_definite, cmap='coolwarm')
ax3.set_title('Semi-Definite A')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('f(x)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# تعریف بردارهای x و y در فضای دو بعدی و سه بعدی
x_values = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_values, x_values)

# فضای دو بعدی (n=2)
A_positive_2d = np.array([[2, 0], [0, 3]])  # ماتریس قطری با درایه‌های مثبت
A_negative_2d = np.array([[-2, 0], [0, -3]])  # ماتریس قطری با درایه‌های منفی

# فضای سه بعدی (n=3)
# برای رسم در فضای سه بعدی باید تابع را ساده‌تر کنیم
A_positive_3d = np.diag([2, 3, 4])  # ماتریس قطری با درایه‌های مثبت
A_negative_3d = np.diag([-2, -3, -4])  # ماتریس قطری با درایه‌های منفی

# تابع f(x) = x^T A x برای فضای دو بعدی
def quadratic_form_2d(A, X, Y):
    return A[0, 0] * X**2 + A[1, 1] * Y**2

# تابع f(x) = x^T A x برای فضای سه بعدی
def quadratic_form_3d(A, x, y, z):
    return A[0, 0] * x**2 + A[1, 1] * y**2 + A[2, 2] * z**2

# محاسبه خروجی تابع برای فضای دو بعدی
Z_positive_2d = quadratic_form_2d(A_positive_2d, X, Y)
Z_negative_2d = quadratic_form_2d(A_negative_2d, X, Y)

# رسم نمودار برای فضای دو بعدی
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# نمودار برای ماتریس قطری با درایه‌های مثبت (n=2)
axs[0].contourf(X, Y, Z_positive_2d, cmap='viridis')
axs[0].set_title('Positive Diagonal Matrix (n=2)')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# نمودار برای ماتریس قطری با درایه‌های منفی (n=2)
axs[1].contourf(X, Y, Z_negative_2d, cmap='plasma')
axs[1].set_title('Negative Diagonal Matrix (n=2)')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.show()

# رسم برای فضای سه بعدی
fig = plt.figure(figsize=(12, 6))

# فضای سه بعدی و بردارهای x, y, z
x_values_3d = np.linspace(-10, 10, 50)
y_values_3d = np.linspace(-10, 10, 50)
z_values_3d = np.linspace(-10, 10, 50)

X_3d, Y_3d, Z_3d = np.meshgrid(x_values_3d, y_values_3d, z_values_3d)

# محاسبه خروجی تابع برای فضای سه بعدی با ماتریس مثبت و منفی
def plot_3d(A, title):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    x = X_3d.flatten()
    y = Y_3d.flatten()
    z = Z_3d.flatten()
    f_values = quadratic_form_3d(A, x, y, z)
    ax.scatter(x, y, z, c=f_values, cmap='coolwarm', s=5)
    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()

# رسم نمودار برای فضای سه بعدی با ماتریس قطری مثبت
plot_3d(A_positive_3d, 'Positive Diagonal Matrix (n=3)')

# رسم نمودار برای فضای سه بعدی با ماتریس قطری منفی
plot_3d(A_negative_3d, 'Negative Diagonal Matrix (n=3)')
```
