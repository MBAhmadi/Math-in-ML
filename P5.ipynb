```python
import numpy as np
import matplotlib.pyplot as plt

lambda1 = 3
lambda2 = 2

x1 = np.linspace(-10, 10, 500)
x2 = np.linspace(-10, 10, 500)
X1, X2 = np.meshgrid(x1, x2)
f = lambda1 * X1**2 + lambda2 * X2**2

plt.contour(X1, X2, f, levels=10)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Ellipse for f(x) in 2D')
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

lambda1 = 3
lambda2 = 2
lambda3 = 1

x1 = np.linspace(-10, 10, 50)
x2 = np.linspace(-10, 10, 50)
x3 = np.linspace(-10, 10, 50)
X1, X2, X3 = np.meshgrid(x1, x2, x3)
f = lambda1 * X1**2 + lambda2 * X2**2 + lambda3 * X3**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X1[:,:,0], X2[:,:,0], f[:,:,0], 50)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x)')
plt.show()
```
