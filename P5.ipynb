```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#   x  y       
x_values = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_values, x_values)

#    (n=2)
A_positive_2d = np.array([[2, 0], [0, 3]])  #     
A_negative_2d = np.array([[-2, 0], [0, -3]])  #     

#    (n=3)
#           
A_positive_3d = np.diag([2, 3, 4])  #     
A_negative_3d = np.diag([-2, -3, -4])  #     

#  f(x) = x^T A x    
def quadratic_form_2d(A, X, Y):
    return A[0, 0] * X**2 + A[1, 1] * Y**2

#  f(x) = x^T A x    
def quadratic_form_3d(A, x, y, z):
    return A[0, 0] * x**2 + A[1, 1] * y**2 + A[2, 2] * z**2

#       
Z_positive_2d = quadratic_form_2d(A_positive_2d, X, Y)
Z_negative_2d = quadratic_form_2d(A_negative_2d, X, Y)

#      
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

#        (n=2)
axs[0].contourf(X, Y, Z_positive_2d, cmap='viridis')
axs[0].set_title('Positive Diagonal Matrix (n=2)')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

#        (n=2)
axs[1].contourf(X, Y, Z_negative_2d, cmap='plasma')
axs[1].set_title('Negative Diagonal Matrix (n=2)')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.show()

#     
fig = plt.figure(figsize=(12, 6))

#      x, y, z
x_values_3d = np.linspace(-10, 10, 50)
y_values_3d = np.linspace(-10, 10, 50)
z_values_3d = np.linspace(-10, 10, 50)

X_3d, Y_3d, Z_3d = np.meshgrid(x_values_3d, y_values_3d, z_values_3d)

#            
def plot_3d(A, title):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    x = X_3d.flatten()
    y = Y_3d.flatten()
    z = Z_3d.flatten()
    f_values = quadratic_form_3d(A, x, y, z)
    ax.scatter(x, y, z, c=f_values, cmap='coolwarm', s=5)
    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()

#          
plot_3d(A_positive_3d, 'Positive Diagonal Matrix (n=3)')

#          
plot_3d(A_negative_3d, 'Negative Diagonal Matrix (n=3)')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define x and y vectors for the 2D case
x_values = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_values, x_values)

# Define diagonal matrix for 2D
A_positive_2d = np.array([[2, 0], [0, 3]])  # Positive diagonal matrix
A_negative_2d = np.array([[-2, 0], [0, -3]])  # Negative diagonal matrix

# Function f(x) for 2D
def f_2d(A, X, Y):
    return A[0, 0] * X**2 + A[1, 1] * Y**2

# Compute f(x) for positive and negative diagonal matrices
Z_positive_2d = f_2d(A_positive_2d, X, Y)
Z_negative_2d = f_2d(A_negative_2d, X, Y)

# Contour level for f(x) = 1
contour_level = 1

# Plot 2D contour plots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Positive diagonal matrix contour
axs[0].contour(X, Y, Z_positive_2d, levels=[contour_level], colors='blue')
axs[0].set_title('Positive Diagonal Matrix (n=2), f(x) = 1')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Negative diagonal matrix contour
axs[1].contour(X, Y, Z_negative_2d, levels=[contour_level], colors='red')
axs[1].set_title('Negative Diagonal Matrix (n=2), f(x) = 1')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')

plt.show()

# 3D Plotting
fig = plt.figure(figsize=(12, 6))

# Define grid for 3D space
x_values_3d = np.linspace(-5, 5, 100)
y_values_3d = np.linspace(-5, 5, 100)
X_3d, Y_3d = np.meshgrid(x_values_3d, y_values_3d)

# Define diagonal matrix for 3D
A_positive_3d = np.diag([2, 3, 4])  # Positive diagonal matrix
A_negative_3d = np.diag([-2, -3, -4])  # Negative diagonal matrix

# Function f(x) for 3D (Z as a function of X and Y)
def f_3d(A, X, Y):
    Z = 1 - (A[0, 0] * X**2 + A[1, 1] * Y**2) / A[2, 2]
    Z = np.where(Z < 0, np.nan, np.sqrt(Z))  # Handle negative values by setting them to NaN
    return Z

# Compute f(x) for positive and negative diagonal matrices in 3D
Z_positive_3d = f_3d(A_positive_3d, X_3d, Y_3d)
Z_negative_3d = f_3d(A_negative_3d, X_3d, Y_3d)

# 3D plot for positive diagonal matrix
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_3d, Y_3d, Z_positive_3d, cmap='coolwarm', edgecolor='none')
ax.set_title('Positive Diagonal Matrix (n=3)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()

# 3D plot for negative diagonal matrix
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_3d, Y_3d, Z_negative_3d, cmap='coolwarm', edgecolor='none')
ax.set_title('Negative Diagonal Matrix (n=3)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
