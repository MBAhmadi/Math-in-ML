```python
import numpy as np
import matplotlib.pyplot as plt

# Define the transformation matrix
A = np.array([[1, 0.7],
              [0, 1]])

# Original square vertices (clockwise starting from the bottom-left corner)
square = np.array([[0, 0],
                   [1, 0],
                   [1, 1],
                   [0, 1],
                   [0, 0]])  # Closing the square by repeating the first vertex

# Transform the square vertices using matrix A
transformed_square = square @ A.T

# Plotting the original and transformed square
plt.figure(figsize=(8, 8))
plt.plot(square[:, 0], square[:, 1], 'bo-', label="Original Square")
plt.plot(transformed_square[:, 0], transformed_square[:, 1], 'ro-', label="Transformed Square")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Original and Transformed Square")
plt.show()

transformed_square
```
